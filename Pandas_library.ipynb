```python
data=[1,2,3,4]
new_data=pd.Series(data)
print(new_data)
```

```python
#custom index
marks=[67,98,78,45,99,98]
subjects=['Urdu','math','Hindi','Os','science','python']
```

```python
pd.Series(marks,index=subjects)
```

```python
pd.Series(marks,index=subjects,name='Marks of Azhar')
```
